# Lightning logs

## FLOP

```python
import matplotlib.pyplot as plt
import pandas as pd

logs = ['./FLOP/lightning_logs/version_0/metrics.csv', './TURN/lightning_logs/version_6/metrics.csv', './RIVER/lightning_logs/version_40/metrics.csv']
for path in logs:
    df = pd.read_csv(path)
    df_train_loss = df[df["train_loss"].notnull()]
    df_val_loss = df[df["val_loss"].notnull()]
    df_accum_train_loss = df_train_loss.groupby('epoch').mean()
    df_accum_val_loss = df_val_loss.groupby('epoch').mean()

    # Create a figure and a set of subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))  # 1 row, 2 columns, and optional figure size

    # Plotting training loss on the first subplot
    ax1.plot(df_accum_train_loss['train_loss'], label='train_loss')
    ax1.set_title('Training Loss for {}'.format(path.split('/')[1]))
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Loss')
    ax1.legend()

    # Plotting validation loss on the second subplot
    ax2.plot(df_accum_val_loss['val_loss'], label='val_loss', color='red')
    ax2.set_title('Validation Loss for {}'.format(path.split('/')[1]))
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Loss')
    ax2.legend()


plt.show()
```
